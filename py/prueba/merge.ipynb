```python
import pandas as pd

# Cargar los archivos CSV
df1 = pd.read_csv('datos_vald_final.csv')
df2 = pd.read_csv('perfiles_vald_20250418_194011.csv')

# Mostrar las primeras filas de ambos dataframes para verificar
print("Datos del primer archivo:")
print(df1.head())
print("\nDatos del segundo archivo:")
print(df2.head())

# Verificar si la columna 'profileId' existe en ambos dataframes
print("\nColumnas en el primer archivo:", df1.columns.tolist())
print("Columnas en el segundo archivo:", df2.columns.tolist())

# Realizar el merge de los dataframes por la columna 'profileId'
df_merged = pd.merge(df1, df2, on='profileId', how='inner')

# Mostrar información sobre el resultado del merge
print("\nInformación del merge:")
print(f"Registros en archivo 1: {len(df1)}")
print(f"Registros en archivo 2: {len(df2)}")
print(f"Registros después del merge: {len(df_merged)}")

# Listar todas las columnas disponibles
print("\nTodas las columnas disponibles:")
print(df_merged.columns.tolist())

# Ordenar columnas según el orden especificado
# Primero las columnas prioritarias
column_order = ['Fecha_Nacimiento', 'Nombre_Completo', 'DNI', 'testTypeName', 'profileId']

# Añadir el resto de las columnas que no están en el orden especificado
remaining_columns = [col for col in df_merged.columns if col not in column_order]
column_order.extend(remaining_columns)

# Reorganizar las columnas (solo incluye las que realmente existen en el DataFrame)
existing_columns = [col for col in column_order if col in df_merged.columns]
df_merged = df_merged[existing_columns]

# Mostrar las primeras filas del dataframe combinado y reorganizado
print("\nPrimeras filas del dataframe combinado con columnas reordenadas:")
print(df_merged.head())

# Guardar el resultado en un nuevo archivo CSV
merged_filename = 'datos_perfiles_merged.csv'
df_merged.to_csv(merged_filename, index=False)
print(f"\nArchivo combinado guardado como: {merged_filename}")
```

```python
def merge_csv_files(file1_path, file2_path, output_path):
    """
    Realiza un merge entre dos archivos CSV por la columna 'profileId' y ordena las columnas
    
    Args:
        file1_path: Ruta al primer archivo CSV
        file2_path: Ruta al segundo archivo CSV
        output_path: Ruta donde guardar el archivo combinado
    
    Returns:
        DataFrame con el resultado del merge
    """
    import pandas as pd
    
    # Cargar los archivos CSV
    df1 = pd.read_csv(file1_path)
    df2 = pd.read_csv(file2_path)
    
    # Realizar el merge de los dataframes por la columna 'profileId'
    df_merged = pd.merge(df1, df2, on='profileId', how='inner')
    
    # Orden específico de columnas (según solicitado)
    column_order = ['Fecha_Nacimiento', 'Nombre_Completo', 'DNI', 'testTypeName', 'profileId']
    
    # Añadir el resto de las columnas que no están en el orden especificado
    remaining_columns = [col for col in df_merged.columns if col not in column_order]
    column_order.extend(remaining_columns)
    
    # Reorganizar las columnas (solo incluye las que realmente existen en el DataFrame)
    existing_columns = [col for col in column_order if col in df_merged.columns]
    df_merged = df_merged[existing_columns]
    
    # Guardar el resultado en un nuevo archivo CSV
    df_merged.to_csv(output_path, index=False)
    
    return df_merged
```

```python
# Uso de la función
result_df = merge_csv_files(
    'datos_vald_final.csv',
    'perfiles_vald_20250418_194011.csv',
    'datos_perfiles_merged.csv'
)

# Ver las primeras filas del resultado
print(result_df.head())
```
