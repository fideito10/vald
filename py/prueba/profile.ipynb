```python
# Nueva celda para extraer perfiles con mejor manejo de errores
import requests
import pandas as pd
import json

def extraer_perfiles(token, tenant_id):
    """
    Extrae perfiles de la API de VALD y los convierte a DataFrame
    """
    # URL y configuraci√≥n
    url = "https://prd-aue-api-externalprofile.valdperformance.com/profiles"
    headers = {
        "Authorization": f"Bearer {token}",
        "Accept": "application/json"
    }
    params = {
        "tenantId": tenant_id
    }
    
    print("üîÑ Consultando perfiles de VALD...")
    
    # Realizar la solicitud
    try:
        response = requests.get(url, headers=headers, params=params)
        print(f"üìä C√≥digo de respuesta: {response.status_code}")
        
        if response.status_code == 200:
            # Intentar decodificar como JSON
            try:
                # Examinar el tipo de la respuesta
                contenido = response.json()
                print(f"Tipo de respuesta: {type(contenido)}")
                
                # Si es una lista (array de perfiles)
                if isinstance(contenido, list):
                    print(f"‚úÖ Se encontraron {len(contenido)} perfiles")
                    
                    # Mostrar los perfiles en formato tabla
                    if len(contenido) > 0:
                        # Extraer informaci√≥n relevante para un DataFrame
                        datos_perfiles = []
                        for perfil in contenido:
                            try:
                                datos_perfiles.append({
                                    'ID': perfil.get('id', 'N/A'),
                                    'Nombre': perfil.get('givenName', 'N/A'),
                                    'Apellido': perfil.get('familyName', 'N/A'),
                                    'Email': perfil.get('email', 'N/A'),
                                    'Sexo': perfil.get('sex', 'N/A'),
                                    'Fecha Nacimiento': perfil.get('dateOfBirth', 'N/A')
                                })
                            except Exception as e:
                                print(f"‚ö†Ô∏è Error procesando un perfil: {str(e)}")
                                print(f"Perfil problem√°tico: {perfil}")
                        
                        # Crear DataFrame y mostrar
                        df_perfiles = pd.DataFrame(datos_perfiles)
                        print("\nüìã Tabla de perfiles:")
                        return df_perfiles
                    else:
                        print("‚ö†Ô∏è No se encontraron perfiles")
                        return pd.DataFrame()
                
                # Si es un objeto (podr√≠a ser un error o un solo perfil)
                elif isinstance(contenido, dict):
                    print("‚ö†Ô∏è La respuesta es un objeto √∫nico, no una lista de perfiles")
                    print(json.dumps(contenido, indent=2))
                    return pd.DataFrame([contenido])
                
                # Otro tipo de respuesta
                else:
                    print(f"‚ö†Ô∏è Formato de respuesta inesperado: {type(contenido)}")
                    print(contenido)
                    return None
                    
            except json.JSONDecodeError:
                print("‚ùå Error decodificando JSON")
                print(f"Respuesta: {response.text[:200]}...")
                return None
                
        elif response.status_code == 401:
            print("üîí Token inv√°lido o vencido")
            return None
        elif response.status_code == 403:
            print("‚õî No tienes permisos suficientes")
            return None
        else:
            print(f"‚ùå Error {response.status_code}")
            print(f"Mensaje: {response.text[:200]}...")
            return None
            
    except Exception as e:
        print(f"‚ùå Error en la solicitud: {str(e)}")
        return None

# Ejecutar la funci√≥n con los par√°metros existentes
df_perfiles = extraer_perfiles(token, tenant_id)
if df_perfiles is not None and not df_perfiles.empty:
    display(df_perfiles)
```

```python
# Funci√≥n simplificada para extraer perfiles de VALD API
import requests
import pandas as pd
import json

def obtener_token():
    """Obtiene un token de autenticaci√≥n para la API de VALD"""
    CLIENT_ID = "Nr37673W7ncT4qBQ=="
    CLIENT_SECRET = "EMiYKddb7wKrwgxxLqopECDpkNLiS1XOEaw="
    TOKEN_URL = "https://security.valdperformance.com/connect/token"
    
    payload = {
        "grant_type": "client_credentials",
        "client_id": CLIENT_ID,
        "client_secret": CLIENT_SECRET
    }
    
    try:
        response = requests.post(TOKEN_URL, data=payload)
        if response.status_code == 200:
            return response.json().get('access_token')
        else:
            print(f"Error obteniendo token: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error en la conexi√≥n: {str(e)}")
        return None

def extraer_perfiles(token=None, tenant_id="f1185650-fb79-44a0-8b4b-b2bf82d28c83"):
    """Extrae perfiles de VALD API y devuelve un DataFrame con los datos"""
    # Si no se proporciona token, intenta obtener uno nuevo
    if not token:
        token = obtener_token()
        if not token:
            return None

    # Configuraci√≥n de la solicitud
    url = "https://prd-aue-api-externalprofile.valdperformance.com/profiles"
    headers = {"Authorization": f"Bearer {token}", "Accept": "application/json"}
    params = {"tenantId": tenant_id}
    
    try:
        # Realizar la solicitud
        response = requests.get(url, headers=headers, params=params)
        
        if response.status_code == 200:
            perfiles = response.json()
            
            # Convertir los perfiles a un DataFrame
            if isinstance(perfiles, list) and perfiles:
                datos = []
                for p in perfiles:
                    datos.append({
                        'ID': p.get('id', 'N/A'),
                        'Nombre': p.get('givenName', 'N/A'),
                        'Apellido': p.get('familyName', 'N/A'),
                        'Email': p.get('email', 'N/A'),
                        'Sexo': p.get('sex', 'N/A')
                    })
                return pd.DataFrame(datos)
            else:
                print("No se encontraron perfiles.")
                return pd.DataFrame()
        else:
            print(f"Error: {response.status_code}")
            print(response.text)
            return None
    except Exception as e:
        print(f"Error en la solicitud: {str(e)}")
        return None

# Ejecutar y mostrar resultados
df = extraer_perfiles()
if df is not None and not df.empty:
    print(f"Se encontraron {len(df)} perfiles:")
    display(df)
```
