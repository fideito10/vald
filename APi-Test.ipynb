````python
import requests
import json
import base64

def probar_conexion_vald():
    """Prueba la conexi√≥n b√°sica a la API de VALD"""
    
    # Configuraci√≥n
    CLIENT_ID = "Nr37673W7ncT4qBQ=="  # Tu ID
    CLIENT_SECRET = "EMiYKddb7wKrwgxxLqopECDpkNLiS1XOEaw="  # Tu Secret
    TOKEN_URL = "https://security.valdperformance.com/connect/token"
    
    # Crear headers con autenticaci√≥n b√°sica
    credentials = f"{CLIENT_ID}:{CLIENT_SECRET}"
    encoded_credentials = base64.b64encode(credentials.encode()).decode()
    
    headers = {
        "Authorization": f"Basic {encoded_credentials}",
        "Content-Type": "application/x-www-form-urlencoded"
    }
    
    # Datos para solicitar el token
    payload = {
        "grant_type": "client_credentials"
    }
    
    try:
        print("\nüîç Informaci√≥n de la petici√≥n:")
        print(f"URL: {TOKEN_URL}")
        print("Headers:", json.dumps(headers, indent=2))
        print("Payload:", json.dumps(payload, indent=2))
        
        # Intento de conexi√≥n
        print("\nüîÑ Intentando conexi√≥n con VALD API...")
        response = requests.post(TOKEN_URL, headers=headers, data=payload)
        
        # Verificar respuesta
        print(f"\nüì° C√≥digo de respuesta: {response.status_code}")
        print(f"üìù Headers de respuesta: {dict(response.headers)}")
        
        if response.status_code == 200:
            token_data = response.json()
            print("\n‚úÖ Conexi√≥n exitosa!")
            print("\nDetalles del token:")
            print(f"- Tipo: {token_data.get('token_type', 'No especificado')}")
            print(f"- Expira en: {token_data.get('expires_in', 'No especificado')} segundos")
            return True
        else:
            print("\n‚ùå Error en la conexi√≥n")
            print(f"Mensaje: {response.text}")
            return False
            
    except Exception as e:
        print(f"\n‚ùå Error inesperado: {str(e)}")
        return False

if __name__ == "__main__":
    probar_conexion_vald()
````

````python
import pandas as pd
import os
from datetime import datetime

def exportar_df_a_csv(df, nombre_base="vald_data", carpeta_destino=None, aplanar_columnas=True, 
                      fecha_en_nombre=True, indice=False, columnas_a_expandir=None):
    """
    Exporta un DataFrame que contiene JSON anidado a un archivo CSV prolijo.
    
    Par√°metros:
    -----------
    df : pandas.DataFrame
        El DataFrame a exportar
    nombre_base : str
        Nombre base para el archivo CSV
    carpeta_destino : str
        Ruta donde guardar el archivo (si es None, usa el directorio actual)
    aplanar_columnas : bool
        Si es True, intenta expandir autom√°ticamente columnas con objetos anidados
    fecha_en_nombre : bool
        Si es True, a√±ade fecha y hora al nombre del archivo
    indice : bool
        Si es True, incluye el √≠ndice del DataFrame en el CSV
    columnas_a_expandir : list
        Lista de nombres de columnas espec√≠ficas a expandir manualmente
        
    Retorna:
    --------
    str: Ruta completa del archivo CSV generado
    """
    # Crear una copia del DataFrame para no modificar el original
    df_procesado = df.copy()
    
    # Si carpeta_destino no existe, crearla
    if carpeta_destino is not None:
        os.makedirs(carpeta_destino, exist_ok=True)
    else:
        carpeta_destino = ""
        
    # Aplanar columnas con JSON anidado
    columnas_expandidas = []
    
    # Expansi√≥n autom√°tica de columnas anidadas si se solicita
    if aplanar_columnas:
        for columna in df_procesado.columns:
            # Verificar si la columna contiene diccionarios o listas
            if df_procesado[columna].dtype == 'object' and len(df_procesado) > 0:
                primer_valor = df_procesado[columna].dropna().iloc[0] if len(df_procesado[columna].dropna()) > 0 else None
                
                if isinstance(primer_valor, dict):
                    # Expandir diccionarios a columnas
                    print(f"Expandiendo columna anidada: {columna}")
                    columnas_expandidas.append(columna)
                    
                    # Usar json_normalize para extraer campos del diccionario
                    temp_df = pd.json_normalize(df_procesado[columna].tolist())
                    
                    # A√±adir cada campo como una nueva columna con prefijo
                    for subcol in temp_df.columns:
                        df_procesado[f"{columna}_{subcol}"] = temp_df[subcol]
                    
                    # Opcional: eliminar la columna original con el diccionario
                    # df_procesado.drop(columns=[columna], inplace=True)
                
                elif isinstance(primer_valor, list) and len(primer_valor) > 0:
                    # Para listas simples, podr√≠a intentar concatenarlas en una cadena
                    if all(isinstance(item, (str, int, float)) for item in primer_valor):
                        print(f"Convirtiendo lista en columna {columna} a texto")
                        df_procesado[columna] = df_procesado[columna].apply(
                            lambda x: "|".join(str(i) for i in x) if isinstance(x, list) else x
                        )
    
    # Expansi√≥n manual de columnas espec√≠ficas
    if columnas_a_expandir:
        for columna in columnas_a_expandir:
            if columna in df_procesado.columns and columna not in columnas_expandidas:
                try:
                    print(f"Expandiendo columna solicitada: {columna}")
                    # Similar al proceso anterior pero para columnas espec√≠ficas
                    if df_procesado[columna].dtype == 'object':
                        valores_no_nulos = df_procesado[columna].dropna()
                        if len(valores_no_nulos) > 0:
                            primer_valor = valores_no_nulos.iloc[0]
                            
                            if isinstance(primer_valor, dict):
                                temp_df = pd.json_normalize(df_procesado[columna].tolist())
                                for subcol in temp_df.columns:
                                    df_procesado[f"{columna}_{subcol}"] = temp_df[subcol]
                except Exception as e:
                    print(f"Error al expandir columna {columna}: {str(e)}")
    
    # Formatear nombre de archivo con fecha si se solicita
    if fecha_en_nombre:
        fecha_hora = datetime.now().strftime("%Y%m%d_%H%M%S")
        nombre_archivo = f"{nombre_base}_{fecha_hora}.csv"
    else:
        nombre_archivo = f"{nombre_base}.csv"
    
    # Ruta completa del archivo
    ruta_completa = os.path.join(carpeta_destino, nombre_archivo)
    
    # Exportar a CSV
    df_procesado.to_csv(ruta_completa, index=indice, encoding='utf-8')
    
    print(f"‚úÖ Archivo CSV generado correctamente: {ruta_completa}")
    print(f"üìä Dimensiones: {df_procesado.shape[0]} filas x {df_procesado.shape[1]} columnas")
    
    return ruta_completa

# Ejemplo de uso:
# ruta_csv = exportar_df_a_csv(df, nombre_base="vald_tests", carpeta_destino="datos_exportados")
````

````python
import json
import pandas as pd
import os
from datetime import datetime
from collections import defaultdict

def leer_json_por_linea(ruta_archivo):
    """Lee un archivo JSON que contiene un objeto JSON por l√≠nea.
    
    Args:
        ruta_archivo (str): Ruta al archivo JSON.
        
    Returns:
        list: Lista de objetos JSON.
    """
    datos = []
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            for linea in archivo:
                linea = linea.strip()
                if linea:  # Omitir l√≠neas vac√≠as
                    try:
                        objeto_json = json.loads(linea)
                        datos.append(objeto_json)
                    except json.JSONDecodeError as e:
                        print(f"Error al decodificar l√≠nea: {e}")
                        print(f"L√≠nea problem√°tica: {linea[:100]}...")
        print(f"‚úÖ Se leyeron {len(datos)} objetos JSON del archivo")
    except Exception as e:
        print(f"‚ùå Error al leer el archivo: {e}")
    
    return datos

def aplanar_diccionario(diccionario, prefijo=''):
    """Aplana un diccionario anidado a un nivel.
    
    Args:
        diccionario (dict): Diccionario para aplanar.
        prefijo (str): Prefijo para claves anidadas.
        
    Returns:
        dict: Diccionario aplanado.
    """
    items = []
    
    for clave, valor in diccionario.items():
        nueva_clave = f"{prefijo}_{clave}" if prefijo else clave
        
        if isinstance(valor, dict):
            # Si es un diccionario, aplanar recursivamente
            items.extend(aplanar_diccionario(valor, nueva_clave).items())
        elif isinstance(valor, list):
            # Si es una lista, manejarla seg√∫n su contenido
            if valor and all(isinstance(item, dict) for item in valor):
                # Lista de diccionarios: usar el √≠ndice como parte de la clave
                for i, item in enumerate(valor):
                    items.extend(aplanar_diccionario(item, f"{nueva_clave}_{i}").items())
            else:
                # Lista simple: convertir a string
                if valor:  # Si la lista no est√° vac√≠a
                    items.append((nueva_clave, json.dumps(valor)))
                else:
                    items.append((nueva_clave, ""))
        else:
            # Valor simple
            items.append((nueva_clave, valor))
    
    return dict(items)

def aplanar_lista_diccionarios(lista_diccionarios):
    """Aplana una lista de diccionarios potencialmente anidados.
    
    Args:
        lista_diccionarios (list): Lista de diccionarios para aplanar.
        
    Returns:
        list: Lista de diccionarios aplanados.
    """
    return [aplanar_diccionario(diccionario) for diccionario in lista_diccionarios]

def json_a_dataframe(datos_json, aplanar=True):
    """Convierte datos JSON a un DataFrame de pandas.
    
    Args:
        datos_json (list/dict): Datos JSON para convertir.
        aplanar (bool): Si se deben aplanar estructuras anidadas.
        
    Returns:
        pandas.DataFrame: DataFrame resultante.
    """
    # Si es un solo diccionario, convi√©rtelo a una lista con un elemento
    if isinstance(datos_json, dict):
        datos_json = [datos_json]
    
    # Si est√° vac√≠o, devolver DataFrame vac√≠o
    if not datos_json:
        return pd.DataFrame()
    
    # Aplanar los datos si se solicita
    if aplanar:
        datos_aplanados = aplanar_lista_diccionarios(datos_json)
    else:
        datos_aplanados = datos_json
    
    # Crear DataFrame
    df = pd.DataFrame(datos_aplanados)
    
    print(f"‚úÖ DataFrame creado con {len(df)} filas y {len(df.columns)} columnas")
    return df

def guardar_dataframe_csv(df, nombre_archivo, carpeta_destino=None, fecha_en_nombre=True, indice=False):
    """Guarda un DataFrame en un archivo CSV.
    
    Args:
        df (pandas.DataFrame): DataFrame a guardar.
        nombre_archivo (str): Nombre base del archivo.
        carpeta_destino (str): Carpeta donde guardar el archivo.
        fecha_en_nombre (bool): Si se debe a√±adir fecha al nombre.
        indice (bool): Si se debe incluir el √≠ndice.
        
    Returns:
        str: Ruta del archivo guardado.
    """
    # Crear carpeta destino si no existe
    if carpeta_destino:
        os.makedirs(carpeta_destino, exist_ok=True)
    else:
        carpeta_destino = ""
    
    # A√±adir fecha al nombre si se solicita
    if fecha_en_nombre:
        fecha_hora = datetime.now().strftime("%Y%m%d_%H%M%S")
        nombre_completo = f"{nombre_archivo}_{fecha_hora}.csv"
    else:
        nombre_completo = f"{nombre_archivo}.csv"
    
    # Ruta completa
    ruta_completa = os.path.join(carpeta_destino, nombre_completo)
    
    # Guardar DataFrame
    df.to_csv(ruta_completa, index=indice, encoding='utf-8')
    
    print(f"‚úÖ Archivo guardado: {ruta_completa}")
    print(f"üìä Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
    
    return ruta_completa

# Ejemplo de uso completo
def procesar_archivo_json_a_csv(ruta_entrada, nombre_salida, carpeta_destino=None):
    """Procesa un archivo JSON l√≠nea por l√≠nea y lo guarda como CSV.
    
    Args:
        ruta_entrada (str): Ruta del archivo JSON a procesar.
        nombre_salida (str): Nombre base para el archivo CSV resultante.
        carpeta_destino (str): Carpeta donde guardar el CSV.
        
    Returns:
        str: Ruta del archivo CSV generado.
    """
    print(f"üîÑ Procesando el archivo: {ruta_entrada}")
    
    # Leer el archivo JSON
    datos = leer_json_por_linea(ruta_entrada)
    
    if not datos:
        print("‚ùå No se encontraron datos para procesar")
        return None
    
    # Convertir a DataFrame
    df = json_a_dataframe(datos)
    
    # Guardar como CSV
    ruta_csv = guardar_dataframe_csv(df, nombre_salida, carpeta_destino)
    
    return ruta_csv
````

````python
# Funci√≥n avanzada para manejar casos especiales en la estructura JSON

def normalizar_columnas_especificas(df, columnas_a_expandir=None, max_nivel=3):
    """Normaliza columnas espec√≠ficas que contienen JSON o listas anidadas.
    
    Args:
        df (pandas.DataFrame): DataFrame a procesar.
        columnas_a_expandir (list): Lista de columnas a expandir.
        max_nivel (int): M√°ximo nivel de anidamiento a procesar.
        
    Returns:
        pandas.DataFrame: DataFrame con columnas normalizadas.
    """
    df_resultado = df.copy()
    
    # Si no se especifican columnas, buscar autom√°ticamente
    if not columnas_a_expandir:
        columnas_a_expandir = []
        for col in df.columns:
            # Verificar si contiene JSON o listas anidadas
            if df[col].dtype == 'object' and df[col].notna().any():
                muestra = df[col].dropna().iloc[0]
                if isinstance(muestra, (dict, list, str)):
                    # Si es string, intentar decodificar como JSON
                    if isinstance(muestra, str):
                        try:
                            json_obj = json.loads(muestra)
                            if isinstance(json_obj, (dict, list)):
                                columnas_a_expandir.append(col)
                        except (json.JSONDecodeError, TypeError):
                            pass
                    else:
                        columnas_a_expandir.append(col)
    
    # Procesar cada columna seleccionada
    for nivel in range(max_nivel):
        for col in columnas_a_expandir.copy():  # Usar copia para poder modificar durante el bucle
            if col in df_resultado.columns:
                try:
                    # Para columnas con strings que son JSON
                    if df_resultado[col].dtype == 'object' and isinstance(df_resultado[col].iloc[0], str):
                        try:
                            # Intentar convertir strings a objetos JSON
                            df_resultado[col] = df_resultado[col].apply(
                                lambda x: json.loads(x) if isinstance(x, str) and x.strip() else x
                            )
                        except (json.JSONDecodeError, TypeError):
                            pass
                    
                    # Normalizar diccionarios
                    if df_resultado[col].apply(lambda x: isinstance(x, dict)).any():
                        # Usar json_normalize para expandir diccionarios
                        dict_norm = pd.json_normalize(df_resultado[col].dropna().tolist())
                        
                        # A√±adir nuevas columnas con prefijo
                        for subcol in dict_norm.columns:
                            nueva_col = f"{col}_{subcol}"
                            # Mapear valores a los √≠ndices originales
                            mapping = {}
                            for i, (idx, val) in enumerate(df_resultado[col].dropna().items()):
                                if isinstance(val, dict) and subcol in val:
                                    mapping[idx] = val[subcol]
                            
                            # Crear nueva columna
                            if mapping:
                                df_resultado[nueva_col] = df_resultado.index.map(mapping)
                                columnas_a_expandir.append(nueva_col)  # Posible nueva columna a expandir
                        
                        # Opcional: eliminar columna original
                        # df_resultado.drop(columns=[col], inplace=True)
                        columnas_a_expandir.remove(col)  # Ya no necesita ser expandida
                    
                    # Normalizar listas
                    elif df_resultado[col].apply(lambda x: isinstance(x, list)).any():
                        # Expandir listas de forma especial seg√∫n su contenido
                        primer_elemento = next((x for x in df_resultado[col] if isinstance(x, list) and len(x) > 0), None)
                        
                        if primer_elemento and all(isinstance(item, dict) for item in primer_elemento):
                            # Lista de diccionarios - crear columnas para cada elemento
                            max_items = df_resultado[col].apply(lambda x: len(x) if isinstance(x, list) else 0).max()
                            
                            for i in range(min(max_items, 5)):  # Limitar a 5 elementos por defecto
                                # Extraer elemento i de cada lista
                                serie_elemento = df_resultado[col].apply(
                                    lambda x: x[i] if isinstance(x, list) and len(x) > i else None
                                )
                                
                                if serie_elemento.notna().any():
                                    # Normalizar este elemento
                                    dict_norm = pd.json_normalize(serie_elemento.dropna().tolist())
                                    
                                    # A√±adir columnas con prefijo
                                    for subcol in dict_norm.columns:
                                        nueva_col = f"{col}_{i}_{subcol}"
                                        # Mapear valores
                                        mapping = {}
                                        for idx, val in serie_elemento.items():
                                            if isinstance(val, dict) and subcol in val:
                                                mapping[idx] = val[subcol]
                                        
                                        if mapping:
                                            df_resultado[nueva_col] = df_resultado.index.map(mapping)
                        else:
                            # Lista de valores simples - convertir a string separado por pipe
                            df_resultado[col] = df_resultado[col].apply(
                                lambda x: "|".join(str(i) for i in x) if isinstance(x, list) else x
                            )
                        
                        columnas_a_expandir.remove(col)  # Ya procesada
                    
                except Exception as e:
                    print(f"Error al procesar columna '{col}': {e}")
    
    return df_resultado
````
